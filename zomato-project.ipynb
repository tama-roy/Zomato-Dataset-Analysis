```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df= pd.read_csv('/kaggle/input/zomato-bangalore-restaurants/zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

# DATA CLEANING

* Droping Duplicates

```python
df= df.drop(['url','address','phone', 'menu_item', 'dish_liked','reviews_list'], axis= 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace= True)
df.shape
```

# Cleaning Rate Column

```python

df['rate'].unique()
```

# Removing "NEW", "-", "/5" from Rate Column


```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value= str(value).split('/')
        value= value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

# Finding and droping null values in rate column with mean

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace= True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace= True)
df.head()
```

```python
df.info()
```

```python
df.rename(columns= {'approx_cost(for two people)' : 'Cost2plates','listed_in(type)': 'Type'}, inplace= True)
df.head()
```

```python
df['location'].unique()
```

# Keeping one from Listed in city and location

```python
df= df.drop(['listed_in(city)'], axis= 1)
```

```python
df.head()
```

# Removing from Cost2Plates Column

```python
df['Cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
df['Cost2plates']= df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

# Cleaning rest type column

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type2= df['rest_type'].value_counts(ascending = False)
rest_type2
```

```python
resttype_lessthan_1000= rest_type2[rest_type2<1000]
resttype_lessthan_1000
```

```python
def handle_rest_type(value):
    if(value in resttype_lessthan_1000):
        return 'others'
    else:
        return value
df['rest_type']= df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
        
```

```python
df.head()
```

# Cleaning location column

```python
df['location'].value_counts()
```

```python
location2= df['location'].value_counts(ascending = False)
location2
```

```python
location_lessthan_300= location2[location2<300]
location_lessthan_300
```

```python
def handle_location(value):
    if(value in location_lessthan_300):
            return 'others'
    else:
        return value
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

# Cleaning Cuisines


```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df['Type'].value_counts()
```
